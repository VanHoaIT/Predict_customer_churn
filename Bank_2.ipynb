```python
df1.replace('Unknown', np.nan, inplace=True, regex=True)
# Xóa các dòng có chứa NaN
df2 = df1.dropna()
```

```python
# mã hóa theo phương pháp 0 1 2 3 4
# quy ước các biến Existing Customer = 0,Attrited Customer = 1
df2.Attrition_Flag = df2.Attrition_Flag.replace({'Existing Customer':0, 'Attrited Customer':1})
# Female = 1, Male = 0
df2.Gender = df2.Gender.replace({'F': 1, 'M':0})
# Uneducated=0,High School=1, College=2, Graduate=3, Post-Graduate=4, Doctorate=5
df2.Education_Level	= df2.Education_Level.replace({'Uneducated': 0, 'High School': 1, 'College': 2, 'Graduate': 3, 'Post-Graduate': 4, 'Doctorate': 5})
# Single=0, Married=1, Divorced=2
df2.Marital_Status = df2.Marital_Status.replace({'Single': 0, 'Married': 1, 'Divorced': 2})
# Less than $40K'= 0, '$40K - $60K'= 1, '$80K - $120K'= 2, '$60K - $80K'= 3, '$120K +'= 4
df2.Income_Category = df2.Income_Category.replace({'Less than $40K': 0, '$40K - $60K': 1, '$80K - $120K': 2, '$60K - $80K': 3, '$120K +': 4})
# 'Blue'= 0, 'Silver'= 1, 'Gold'= 2, 'Platinum' 3
df2.Card_Category = df2.Card_Category.replace({'Blue': 0, 'Silver': 1, 'Gold': 2, 'Platinum': 3})

df2.drop(columns = ['CLIENTNUM'],inplace=True)
```

```python
df2
```

```python
df_cluster = df2.copy()
```

```python
wcss = []
for i in range(1, 7):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_cluster)
    wcss.append(kmeans.inertia_)

fig = go.Figure()
fig.add_trace(go.Scatter(x=list(range(1, 7)), y=wcss, mode='lines+markers'))

fig.update_layout(
    title='Elbow Method',
    xaxis_title='Number of clusters',
    yaxis_title='WCSS',
    template='plotly_dark'
)

fig.show()
```

```python
# nên dùng cách lấy cụm
# Xác định số lượng cụm
n_clusters = 5

# Chạy thuật toán KMeans
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
kmeans.fit(df_cluster)

# Gán nhãn cụm cho mỗi mẫu dữ liệu
cluster_labels = kmeans.labels_

# Thêm nhãn cụm vào dataframe
df_cluster['Cluster'] = cluster_labels

# In ra thông tin về các cụm
print(df_cluster.groupby('Cluster').mean())
```

```python
# xác định mức độ quan trọng thuộc tính, để đưa các thuộc tính vào mô hình (cho cả phân cụm nên lấy bao nhiểu cụm)
# dựa vào tương quan, để xác định các thuộc tính quan trọng, các thuộc tính có độ tương quan cao nên gộp dùng 1 cái(vd: hạn mức và chi tiêu trung bình)
# nhập thông tin và0 dựa đoán, nhập các thuộc tính quan trọng để cho mô hình dựa đoán
```

```python
# xây dựng trên cả tập dữ liệu
X_train, X_test, y_train, y_test = train_test_split(usampled_df.drop(columns=['Churn']), usampled_df['Churn'], test_size=0.2, random_state=42)
# Khởi tạo mô hình Rừng ngẫu nhiên
rf_model = RandomForestClassifier(random_state=42)

# Huấn luyện mô hình
rf_model.fit(X_train, y_train)

# Lấy độ quan trọng của thuộc tính
feature_importances = rf_model.feature_importances_
feature_names = X_train.columns

# Tạo DataFrame để hiển thị độ quan trọng của thuộc tính
feature_importance_df = pd.DataFrame({'Thuộc tính': feature_names, 'Độ quan trọng': feature_importances})

# Sắp xếp các thuộc tính theo độ quan trọng
feature_importance_df = feature_importance_df.sort_values(by='Độ quan trọng', ascending=False)

# In ra các thuộc tính quan trọng nhất
print(feature_importance_df)

# Chọn top 7 thuộc tính quan trọng nhất
top_features = feature_importance_df['Thuộc tính'].head(7).tolist()
```

```python
top_features
```

```python
# Chỉ lấy các thuộc tính đã chọn
X_train_selected = X_train[top_features]
X_test_selected = X_test[top_features]

# Khởi tạo mô hình Rừng ngẫu nhiên
rf_model2 = RandomForestClassifier(random_state=42)

# Huấn luyện mô hình với các thuộc tính đã chọn
rf_model2.fit(X_train_selected, y_train)

# Dự đoán trên tập kiểm tra
y_pred = rf_model2.predict(X_test_selected)

# Đánh giá mô hình
print(classification_report(y_test, y_pred))
```
